```python
def log(left, right, target, base):
    mid = (left + right) / 2
    cur = base ** mid
    if abs(cur - target) < 0.0000001:
        return mid
    elif right - left < 0.0000001:
        return mid
    elif cur < target:
        return log(mid, right, target, base)
    else:
        return log(left, mid, target, base)
```

```python
log(0, 10, 8, 2)
```
