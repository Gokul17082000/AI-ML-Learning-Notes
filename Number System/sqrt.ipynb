```python
def sqrt(left, right, target):
    mid = (left + right) / 2
    cur = mid * mid
    if abs(cur - target) < 0.0000001:
        return mid
    elif right - left < 0.0000001:
        return mid
    elif cur < target:
        return sqrt(mid, right, target)
    else:
        return sqrt(left, mid, target)
```
