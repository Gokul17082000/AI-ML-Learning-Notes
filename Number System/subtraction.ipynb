```python
def subtract(x, y, base):
    ans = 0
    bal = 0
    place = 1

    while x > 0 or y > 0:
        r1 = x % 10
        r2 = y % 10

        r1 = r1 - bal
        bal = 0

        if r1 < r2:
            bal = 1
            r1 = r1 + base


        cur = r1 - r2
        rem = cur % base

        ans = ans + rem * place

        x = x // 10
        y = y // 10

        place = place * 10

    return ans
```

```python
subtract(12, 7, 8)
```
