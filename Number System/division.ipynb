```python
base = 8
x = 12
y = 7

ans = 0
pow = 1

while y > 0:
    bal = 0
    mul = y % 10

    t = x

    cur = 0
    p = 1
    while t > 0:
        r1 = t % 10
      
        tot = r1 * mul + bal
        rem = tot % base
        bal = tot // base

        cur = cur + rem * p

        p = p * 10
        t = t // 10
    
    if bal > 0:
        cur = cur + bal * p

    ans = ans + cur*pow

    y = y // 10

    pow = pow * 10

print(ans)    

    
```
