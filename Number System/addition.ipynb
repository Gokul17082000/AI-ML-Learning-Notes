```python
symbols = {
    10: "A",
    11: "B",
    12: "C",
    13: "D",
    14: "E",
    15: "F"
}

revsymbols = {v: k for k, v in symbols.items()}

def int_to_char(x):
    if x > 9:
        return symbols[x]
    else:
        return str(x)

def char_to_int(x):
    if x in revsymbols:
        return revsymbols[x]    
    else:
        return int(x)

def add(x, y, base):
    ans = ""
    bal = 0

    i = len(x) - 1
    j = len(y) - 1

    while i >= 0 or j >= 0 or bal > 0:
        r1 = char_to_int(x[i]) if i>=0 else 0
        r2 = char_to_int(y[j]) if j>=0 else 0

        cur = r1 + r2 + bal
        bal = cur // base
        rem = cur % base

        ans = int_to_char(rem) + ans

        i = i - 1
        j = j - 1

    return ans
    
```

```python
add("16", "16", 8)
```

```python
add("A5", "1F", 16)
```
