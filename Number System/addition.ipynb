```python
def add(x, y, base):
    ans = 0
    bal = 0
    place = 1

    while x > 0 or y > 0:
        r1 = x % 10
        r2 = y % 10

        cur = r1 + r2 + bal
        bal = cur // base
        rem = cur % base

        ans = ans + rem * place

        x = x // 10
        y = y // 10

        place = place * 10

    if bal > 0:
        ans = ans + bal * place

    return ans
    
```

```python
add(16, 16, 8)
```
