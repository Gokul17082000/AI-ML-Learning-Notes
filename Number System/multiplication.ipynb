```python
def add(x, y, base):
    ans = 0
    bal = 0
    place = 1

    while x > 0 or y > 0:
        r1 = x % 10
        r2 = y % 10

        cur = r1 + r2 + bal
        bal = cur // base
        rem = cur % base

        ans = ans + rem * place

        x = x // 10
        y = y // 10

        place = place * 10

    if bal > 0:
        ans = ans + bal * place

    return ans
```

```python
def multiply(x, y, base):
    ans = 0
    place_y = 1

    while y > 0:
        bal = 0
        mul = y % 10

        t = x

        cur = 0
        place_x = 1
        while t > 0:
            r1 = t % 10
        
            tot = r1 * mul + bal
            rem = tot % base
            bal = tot // base

            cur = cur + rem * place_x

            place_x = place_x * 10
            t = t // 10
        
        if bal > 0:
            cur = cur + bal * place_x

        ans = add(ans, cur*place_y, base)

        y = y // 10

        place_y = place_y * 10

    return ans 
```

```python
multiply(12, 7, 8)
```
