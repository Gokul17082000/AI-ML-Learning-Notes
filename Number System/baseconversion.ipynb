```python
import string

digits = string.digits
uppercase = string.ascii_uppercase 
lowercase = string.ascii_lowercase
all_chars = digits + uppercase + lowercase

symbols = {i: char for i, char in enumerate(all_chars)}

revsymbols = {v: k for k, v in symbols.items()}

def int_to_char(x):
    return symbols[x]

def char_to_int(x):
    return revsymbols[x]    

def baseconversion(x, base1, base2):

    if base1 < 2 or base1 > len(all_chars):
        raise ValueError("Invalid base1")

    if base2 < 2 or base2 > len(all_chars):
        raise ValueError("Invalid base2")

    isNegative = False
    if x[0] == "-":
        isNegative = True
        x = x[1:]

    decimal = 0
    for ch in x:
        val = char_to_int(ch)
        if val >= base1:
            raise ValueError("Invalid digit for base1")
        decimal = decimal * base1 + val

    if decimal == 0:
        return "0"

    result = ""
    while decimal > 0:
        rem = decimal % base2
        result = int_to_char(rem) + result
        decimal //= base2

    if isNegative:
        result = "-" + result

    return result
```

```python
baseconversion("5", 10, 2)
```
