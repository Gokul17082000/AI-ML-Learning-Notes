```python
def baseconversion(x, base1, base2):
    ans = 0
    place = 1

    t = x
    while t > 0:
        r = t % 10
        ans = ans + place * r
        place = place * base1
        t = t // 10
    
    res = 0
    place = 1

    while ans > 0:
        r = ans % base2
        res = res + place * r
        ans = ans // base2
        place = place * 10

    return res
```

```python
baseconversion(5, 10, 2)
```
