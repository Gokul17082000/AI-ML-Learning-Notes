# If a bag has 80% Red and 20% Black, would its impurity be closer to 0 or 1? Why? 

- Impurity is closer to 0, because the bag is mostly one color (Red), which indicates low impurity. The more one color dominates, the lower the impurity, as there is less diversity in the contents of the bag.

# What is impurity(0.9, 0.1)? Higher or lower than impurity(0.8, 0.2)?

- Impurity(0.9, 0.1) is lower than impurity(0.8, 0.2) because the first bag has a higher concentration of one color (90% Red) compared to the second bag (80% Red). The more skewed the distribution towards one color, the lower the impurity.

# What mathematical property must the function impurity(p, q) have when p = 1 - q?

- The function impurity(p, q) must satisfy the property that impurity(p, q) = impurity(1 - q, 1 - p). This is because the impurity should be symmetric with respect to the two classes (Red and Black). If we swap the proportions of the two classes, the impurity should remain the same, reflecting that it is a measure of how mixed the classes are, regardless of which class is more dominant.

# Is impurity(p, q) the same as impurity(q, p)? Why should it be?

- Yes, impurity(p, q) should be the same as impurity(q, p) because impurity is a measure of how mixed the classes are, and it should not depend on the order of the classes. The function should be symmetric, meaning that swapping the proportions of the two classes should yield the same impurity value. This ensures that the impurity measure is consistent and unbiased towards any particular class.
