```python
import math
```

```python
def find_distance_2d(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def closer_point(p, a, b):
    d1 = find_distance_2d(p[0], p[1], a[0], a[1])
    d2 = find_distance_2d(p[0], p[1], b[0], b[1])

    if d1 < d2:
        return a
    elif d1 > d2:
        return b
    else:
        return "Equal"

print(closer_point((1, 2), (0, 0), (5, 5)))
print(closer_point((4, 4), (0, 0), (8, 8)))
print(closer_point((7, 3), (2, 3), (10, 3)))
```

```python
def is_point_on_line_1d(a, b, p):
    return min(a, b) <= p <= max(a, b)

print(is_point_on_line_1d(2, 5, 3))
print(is_point_on_line_1d(5, 2, 3))
print(is_point_on_line_1d(2, 5, 6))
print(is_point_on_line_1d(4, 4, 4))
print(is_point_on_line_1d(4, 4, 5))
print(is_point_on_line_1d(-5, -2, -3))
print(is_point_on_line_1d(-5, -2, -6))
print(is_point_on_line_1d(-2, -5, -4))
print(is_point_on_line_1d(-1, 3, 0))
```

```python
def is_point_on_line_2d(x1, y1, x2, y2, x, y):
    return (x2 - x1) * (y - y1) == (y2 - y1) * (x - x1)

print(is_point_on_line_2d(0, 0, 3, 4, 1, 2))
print(is_point_on_line_2d(0, 0, 1, 1, 2, 2))
```

```python
def are_lines_touching_or_overlapping(a1, b1, a2, b2):
    a1, b1 = min(a1, b1), max(a1, b1)
    a2, b2 = min(a2, b2), max(a2, b2)
    return max(a1, a2) <= min(b1, b2)

print(are_lines_touching_or_overlapping(1, 4, 3, 6))
print(are_lines_touching_or_overlapping(1, 3, 3, 5))
print(are_lines_touching_or_overlapping(1, 2, 3, 4))
print(are_lines_touching_or_overlapping(-3, 1, 0, 4))
print(are_lines_touching_or_overlapping(-5, -2, -2, 3))
print(are_lines_touching_or_overlapping(-10, -6, -5, -1))
print(are_lines_touching_or_overlapping(-2, -7, -4, -3))
```

```python
def are_lines_touching_or_overlapping_2d(a1, b1, a2, b2):
    return are_lines_touching_or_overlapping(a1[0], a1[1], a2[0], a2[1]) and are_lines_touching_or_overlapping(b1[0], b1[1], b2[0], b2[1])
```

```python
def is_point_inside_rectangle(x1, y1, x2, y2, x, y):
    return x1 <= x <= x2 and y1 <= y <= y2

print(is_point_inside_rectangle(0, 0, 10, 5, 3, 2))
print(is_point_inside_rectangle(0, 0, 10, 5, 10, 5))
print(is_point_inside_rectangle(0, 0, 10, 5, 11, 5))
print(is_point_inside_rectangle(-5, -5, 5, 5, 0, 0))
```

```python
def are_rectangles_intersecting(rect1: tuple, rect2: tuple) -> bool:
    (x1a, y1a), (x2a, y2a) = rect1
    (x1b, y1b), (x2b, y2b) = rect2

    return (
        max(x1a, x1b) <= min(x2a, x2b) and
        max(y1a, y1b) <= min(y2a, y2b)
    )

print(are_rectangles_intersecting(((0, 0), (3, 3)), ((2, 2), (5, 5))))
print(are_rectangles_intersecting(((0, 0), (1, 1)), ((2, 2), (3, 3))))
print(are_rectangles_intersecting(((0, 0), (2, 2)), ((2, 2), (4, 4))))
print(are_rectangles_intersecting(((0, 0), (5, 5)), ((1, 1), (2, 2))))
```

```python
def my_impurity(c1, c2):
    tot = c1 + c2
    return 1 - abs(c1 - c2) / tot

print(my_impurity(0, 5))
print(my_impurity(5, 5))
print(my_impurity(7, 3))
print(my_impurity(9, 1))
```
