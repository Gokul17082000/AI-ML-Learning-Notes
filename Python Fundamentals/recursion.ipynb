```python
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

print(factorial_recursive(5))
print(factorial_recursive(3))
print(factorial_recursive(0))
print(factorial_recursive(1))
```

```python
def multiply_recursive(a, b):
    if b == 0:
        return 0
    if b > 0:
        return a + multiply_recursive(a, b - 1)
    else:
        return -multiply_recursive(a, -b)

print(multiply_recursive(4, 3))
print(multiply_recursive(5, 0))
print(multiply_recursive(7, -2))
print(multiply_recursive(-3, -3)) 
```

```python
def power(x, n):
    if n == 0:
        return 1
    else:
        return x * power(x, n - 1)
    
print(power(2, 3))
print(power(5, 2))
```

```python
def compute_power(x, n):
    if n == 0:
        return 1
    elif n > 0:
        return x * power(x, n - 1)
    else:
        return 1 / power(x, -n)
    
print(compute_power(2, 3))
print(compute_power(2, -3))
print(compute_power(5, 0))   
```

```python
def recursive_divide(dividend, divisor):
    if dividend < divisor:
        return (0, dividend)
    q, r = recursive_divide(dividend - divisor, divisor)
    return (q + 1, r)

print(recursive_divide(17, 5))
print(recursive_divide(20, 4))
print(recursive_divide(7, 3))
print(recursive_divide(0, 1))
```

```python
def compute_hcf(a, b):
    if b == 0:
        return a
    return compute_hcf(b, a % b)

print(compute_hcf(12, 18))
print(compute_hcf(100, 25))
print(compute_hcf(17, 13))
print(compute_hcf(0, 5))
```

```python
def solve_hanoi(n, src="A", aux="B", dest="C"):
    if n==1:
        print("Move {} from {} to {}".format(n, src, dest))
        return
    solve_hanoi(n-1, src, dest, aux)
    print("Move {} from {} to {}".format(n, src, dest)) 
    solve_hanoi(n-1, aux, src, dest)   

print("1 disk:")
solve_hanoi(1)
print("2 disk:")
solve_hanoi(2)
print("3 disk:")
solve_hanoi(3)
```

```python
def permutation(num):
    ans = []
    for i in range(len(num)):
        for j in range(len(num)):
            num[i], num[j] = num[j], num[i]
            ans.append("".join(num))
            
    return set(ans)

s = "pri"
print(permutation(list(s)))
s = "123"
print(permutation(list(s)))
s = "1234"
print(permutation(list(s)))
```
