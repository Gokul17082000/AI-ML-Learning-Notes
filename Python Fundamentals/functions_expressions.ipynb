```python
def extract_last_digit(n):
    return n % 10

print(extract_last_digit(123))
print(extract_last_digit(0))
```

```python
def remove_last_digit(n):
    return n // 10

print(remove_last_digit(123))
print(remove_last_digit(0))
```

```python
def calculate_interest(principal, rate, time):
    interest = ( principal * rate * time ) / 100
    return interest

print(calculate_interest(1000, 5, 2))
print(calculate_interest(1500, 4.3, 3))
print(calculate_interest(500, 10, 0))
```

```python
import math
```

```python
def calculate_hypotenuse(base, height):
    return math.sqrt(base**2 + height**2)

print(calculate_hypotenuse(3, 4))
print(calculate_hypotenuse(5, 12))
print(calculate_hypotenuse(0, 0))
```

```python
def find_distance_2d(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print(find_distance_2d(0, 0, 3, 4))
print(find_distance_2d(1, 2, 4, 6))
```

```python
def calculate_distance_3d(x1, y1, z1, x2, y2, z2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)

print(calculate_distance_3d(0, 0, 0, 1, 1, 1))
print(calculate_distance_3d(2, 3, 5, 2, 3, 5))
print(calculate_distance_3d(1, 2, 3, 4, 6, 9))
```

```python
def manhattan_distance(x1, y1, x2, y2):
    return abs(x1-x2) + abs(y1-y2)

print(manhattan_distance(2, 3, 5, 1))
print(manhattan_distance(0, 0, 0, 0))
print(manhattan_distance(-1, -1, 1, 1))
```

```python
def predict(m, c, x):
    return m * x + c

print(predict(2, 3, 4))
print(predict(-1, 5, 2))
```

```python
def fit(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    c = y1 - m * x1
    return m, c

print(fit(1, 2, 3, 6))
print(fit(0, 5, 2, 9))
```

```python
def approx_derivative(func, x, h):
    return (func(x + h) - func(x)) / h

print(approx_derivative(lambda x: x*x, 3, 0.001))
print(approx_derivative(lambda x: x*x*x, 2, 0.0001))
print(approx_derivative(lambda x: math.sin(x), 2, 0.0001))
print(approx_derivative(lambda x: math.log(x), 2, 0.0001))
print(approx_derivative(lambda x: math.exp(x), 2, 0.0001))
```
