```python
def sqrt_binary_search(n, precision):
    if n < 0:
        return None

    low, high = 0, max(1, n)

    while high - low > precision:
        mid = (low + high) / 2
        if mid * mid > n:
            high = mid
        else:
            low = mid

    return (low + high) / 2
    
print(sqrt_binary_search(25, 0.0001))
print(sqrt_binary_search(2, 0.0001))
```

```python
def cuberoot_binary_search(n, precision):    
    
    sign = 1
    if n < 0:
        sign = -1
        n = -n

    low, high = 0, max(1, n)

    while high - low > precision:
        mid = (low + high) / 2
        if mid * mid * mid > n:
            high = mid
        else:
            low = mid

    return sign * (low + high) / 2

print(cuberoot_binary_search(27, 1e-6))
print(cuberoot_binary_search(8, 1e-6))
print(cuberoot_binary_search(-64, 1e-6))
print(cuberoot_binary_search(0.001, 1e-6))
```

```python
def compute_nth_root(n, k, precision=1e-6):
    if n == 0:
        return 0

    sign = 1
    if n < 0:
        if k % 2 == 0:
            return None
        sign = -1
        n = -n

    low, high = 0, max(1, n)

    while high - low > precision:
        mid = (low + high) / 2
        if mid ** k > n:
            high = mid
        else:
            low = mid

    return sign * (low + high) / 2

print(compute_nth_root(16, 2))
print(compute_nth_root(27, 3))
print(compute_nth_root(81, 4))
```

```python
def log10_binary_search(x, precision=1e-7):
    if x <= 0:
        return None

    low, high = 0, max(1, x)

    if x < 1:
        low, high = -1, 0
        while 10 ** low > x:
            high = low
            low -= 1

    while high - low > precision:
        mid = (low + high) / 2
        try:
            if 10 ** mid > x:
                high = mid
            else:
                low = mid
        except OverflowError:
            high = mid 

    return (low + high) / 2

print(log10_binary_search(100))
print(log10_binary_search(1000))
print(log10_binary_search(0.01))
print(log10_binary_search(2), 5)
```

```python
def log_base_n(x, n, precision=1e-7):
    if x <= 0:
        return None

    low, high = 0, max(1, x)

    if x < 1:
        low, high = -1, 0
        while n ** low > x:
            high = low
            low -= 1

    while high - low > precision:
        mid = (low + high) / 2
        try:
            if n ** mid > x:
                high = mid
            else:
                low = mid
        except OverflowError:
            high = mid 

    return (low + high) / 2

print(log_base_n(8, 2))
print(log_base_n(81, 3))
print(log_base_n(0.04, 5))
print(log_base_n(10, 2), 5)
```
