```python
def word_count(s):
    cnt = dict()
    token = s.split()
    for i in token:
        cnt[i] = cnt.get(i, 0) + 1
    return cnt

print(word_count("hello world hello"))
print(word_count("apple orange apple banana orange"))
```

```python
def find_anagrams(s):
    tokens = s.split()
    anagrams = dict()
    for i in tokens:
        key = ''.join(sorted(i))
        anagrams[key] = anagrams.get(key, []) + [i]
    return anagrams

print(find_anagrams("listen silent enlist inlets hello world"))
print(find_anagrams("evil vile live veil")
)
```

```python
def update_expenses(expense_list, expenses_dict):
    for i in expense_list:
        expenses_dict[i[0]] = expenses_dict.get(i[0], 0) + i[1]    

def print_expenses(expenses_dict):
    for k, v in sorted(expenses_dict.items()):
        print(k, v)

expenses = {}

update_expenses([("food", 200), ("rent", 1000)], expenses)
update_expenses([("food", 300), ("travel", 150)], expenses)

print_expenses(expenses)
```
